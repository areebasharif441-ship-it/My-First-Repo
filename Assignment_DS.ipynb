```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/sample_data/ecommerce_sales.csv")
```

```python
df_shape=df.shape
print(df.shape)
```

```python
df_describe = df.describe()

unique_categories_count = df["category"].nunique()

print(f"Shape: {df_shape}")
print(f"Unique categories: {unique_categories_count}")
display(df_describe)
```

```python
has_missing = df.isnull().any(axis=1)

df["discount_applied"] = df["discount_applied"].fillna(0.0)
print(has_missing)
```

```python
price_arr = df["price"].to_numpy()
quantity_arr = df["quantity"].to_numpy()

gross_cost_array = price_arr * quantity_arr

print(gross_cost_array[:5])
```

```python
df["net_revenue"] = (
    (df["price"] * df["quantity"]) *
    (1 - df["discount_applied"]) +
    df["shipping_cost"]
)

df[["price", "quantity", "discount_applied", "shipping_cost", "net_revenue"]].head()
```

```python
category_stats = df.groupby("category").agg(
    total_net_revenue=("net_revenue", "sum"),
    avg_discount=("discount_applied", "mean")
)
print(category_stats)
```

```python
payment_pivot = pd.pivot_table(
    df,
    index="category",
    columns="payment_method",
    values="net_revenue",
    aggfunc="sum",
    fill_value=0
)
print(payment_pivot)
```

```python
premium_orders = df[
    (df["shipping_cost"] > 10.0) &
    (df["net_revenue"] > 50.0)
].sort_values(by="net_revenue", ascending=False).head(5)
print(premium_orders)
```
