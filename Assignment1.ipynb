```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/sample_data/properties.csv")
```

```python
df.head()
```

```python
missing_buyer_count = df['buyer'].isnull().sum()
buyer_has_nulls = df['buyer'].isnull().any()

print(missing_buyer_count, buyer_has_nulls)
```

```python
df_filled_price = df.copy()
df_filled_price['advertised_price'] = df_filled_price['advertised_price'].fillna(
    df_filled_price['advertised_price'].median()
)

df_filled_price
```

```python
filtered_df = df_filled_price[
    (df_filled_price['location'].isin(['Gulberg', 'DHA'])) &
    (df_filled_price['property_category'].notnull()) &
    (df_filled_price['advertised_price'] > 5000000)
]

filtered_avg_price = filtered_df['advertised_price'].mean()

print(filtered_avg_price)
filtered_df
```

```python
first_remark = str(df_filled_price['remarks'].iloc[0]).lower()
first_good_count = first_remark.count('good')

print(first_good_count)
```

```python
df_filled_price['discount_amount'] = (
    df_filled_price['advertised_price'] - df_filled_price['sale_price']
)

avg_discount = df_filled_price['discount_amount'].mean()

print(avg_discount)
df_filled_price[['advertised_price', 'sale_price', 'discount_amount']].head()
```

```python
filtered_df.to_csv("filtered_properties.csv", index=False)
```
