## Task 3: NumPy Array Operations
Extract the `price` and `quantity` columns as NumPy arrays.
Compute the total cost before discount (`price` multiplied by `quantity`) for each row using NumPy array multiplication, and store the resulting array in `gross_cost_array`.

```python
# TODO
gross_cost_array = df['price'].to_numpy() * df['quantity'].to_numpy()
print(gross_cost_array[:5])
```

## Task 4: Feature Engineering
Create a new column in `df` called `net_revenue`.
The formula is: `(price * quantity) * (1 - discount_applied) + shipping_cost`

```python
# TODO
# TODO
df['net_revenue'] = (df['price'] * df['quantity']) * (1 - df['discount_applied']) + df['shipping_cost']
df[['price', 'quantity', 'discount_applied', 'shipping_cost', 'net_revenue']].head()
```

## Task 5: GroupBy Analysis
Group the data by `category`.
Calculate the **sum** of `net_revenue` and the **mean** of `discount_applied` for each category.
Store the resulting DataFrame in `category_stats`.

```python
# TODO
category_stats = df.groupby('category').agg({
    'net_revenue': 'sum',
    'discount_applied': 'mean'
})

category_stats
```

## Task 6: Pivot Table
Create a pivot table showing the total (sum) `net_revenue` for each `category` (index/rows) broken down by `payment_method` (columns).
Store the pivot table DataFrame in `payment_pivot`. Fill any missing values in the pivot table with `0`.

```python
# TODO
payment_pivot = pd.pivot_table(
    df,
    values='net_revenue',
    index='category',
    columns='payment_method',
    aggfunc='sum',
    fill_value=0
)

payment_pivot
```

## Task 7: Boolean Indexing & Sorting
Find all orders where `shipping_cost` is greater than 10.0 **AND** `net_revenue` is greater than 50.0.
Sort these orders by `net_revenue` in **descending** order.
Store the top 5 resulting rows in a new DataFrame called `premium_orders`.

```python
premium_orders = df[
    (df['shipping_cost'] > 10.0) &
    (df['net_revenue'] > 50.0)
].sort_values(by='net_revenue', ascending=False).head(5)

premium_orders
```
